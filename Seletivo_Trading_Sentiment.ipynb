# Seletivo - Trading Sentiment

```python
# Bibliotecas
import requests
import pandas as pd
import json
import datetime, time
from datetime import timedelta
```

```python
# Informações
symbol = 'BTCUSDT'
period ='5m'
loop = True
conta = 29
pasta = False

# Solicitação start_time e end_time, a entrada fornecida deve estar em milissegundos e em Unix Timestamp
def TimeStamp_Unix(dias=0):
    secunds = datetime.datetime.now() - timedelta(days=dias)
    timeStamp = int(time.mktime(secunds. timetuple()) * 1000)
    return timeStamp
```

```python
#url = ('https://fapi.binance.com/futures/data/globalLongShortAccountRatio?symbol=BTCUSDT&period=5m&startTime=1648245600000&endTime=1650924000000')
def database(start_time, end_time):
  # Buscar dados
  url = requests.get(f"https://fapi.binance.com/futures/data/globalLongShortAccountRatio?symbol={symbol}&period={period}&endTime={end_time}&startTime={start_time}")
  data = json.loads(url.content)
  data = [{'timestamp':selecao['timestamp'],'symbol': selecao['symbol'], 'longShortRatio':selecao['longShortRatio']} for selecao in data]
  data = pd.DataFrame(data)

  # Dataframe para cada chamada e concatená-lo para um final. Verifique pd.concat()
  # Junçaõ de busca
  archive = pd.read_csv('/content/arquivo.csv')
  concat = pd.concat([archive, data])
  concat.drop(concat.columns[concat.columns.str.contains('unnamed')], axis=1, inplace = True)
  concat.to_csv('arquivo.csv', index = False, encoding = 'utf-8')

# Loop de busca
while loop:
  start_time = TimeStamp_Unix(conta)
  end_time = TimeStamp_Unix(conta-1)
  if conta <= 1:
    loop = False
  if pasta != False:
    database(start_time, end_time)
  else:
    url = requests.get(f"https://fapi.binance.com/futures/data/globalLongShortAccountRatio?symbol=BTCUSDT&period=5m&endTime={end_time}&startTime={start_time}")
    data = json.loads(url.content)
    data = [{'timestamp':selecao['timestamp'],'symbol': selecao['symbol'], 'longShortRatio':selecao['longShortRatio']} for selecao in data]
    df = pd.DataFrame(data)
    df.to_csv('arquivo.csv', index = False, encoding = 'utf-8')
    pasta = "/content/arquivo.csv"
  conta -=1
```
